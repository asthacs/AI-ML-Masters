```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Importing the dataset
X_train = pd.read_csv('Linear_X_Train.csv')
y_train = pd.read_csv('Linear_Y_Train.csv')
X_test = pd.read_csv('Linear_X_Test.csv')

# Building and fitting the simple linear regression model
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)

# Predict the values of y
y_pred = regression_model.predict(X_test)
print("Predicted values of y:", y_pred.reshape(1250))

# Plotting Train set
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regression_model.predict(X_train), color='blue')
plt.title('Train Set')
plt.xlabel('Time spent on coding daily')
plt.ylabel('Score in exam')
plt.show()

# Plotting Test set
plt.scatter(X_test, y_pred, color='red')
plt.plot(X_train, regression_model.predict(X_train), color='blue')
plt.title('Test Set')
plt.xlabel('Time spent on coding daily')
plt.ylabel('Score in exam')
plt.show()
```
